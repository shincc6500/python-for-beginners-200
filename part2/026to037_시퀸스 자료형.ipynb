## 26. 시퀸스 자료형 이해하기
시퀸스 자료형은 어떤 객체가 순서를 가지고 나열되어 있는 자료형  
반복문 등을 통해 순서대로 자료 내부의 객체를 사용할 수 있음. 
예시로는 문자열, 리스트, 튜플이 있음. 

문자열은 기호나 문자가 순서대로 나열되어 있는 자료형이며  
리스트와 튜플은 임의의 객체가 나열되어 있는 자료형이다.   
단 리스트는 수정 가능하지만 튜플을 수정이 불가능하다는 특징이 있음.   

파이썬의 시퀸스 자료형은 다음과 같은 공통적인 특성을 가짐. 
|특성|설명|
|-----|-----|
|인덱싱| 인덱스를 통해 해당 값에 접근 가능함. 인덱스는 0부터 시작|
|슬라이싱| 특정 구간의 값을 지정하여 사용 가능함. <br>구간은 [start : end : step]와 같이 시작 인덱스, 끝 인덱스, 간격으로 정의되며 간격은 생략할 시 1을 사용함|
|연결 | + 연산자를 이용하여 두 시퀸스 자료를 연결하여 새로운 시퀸스 자료를 생성함|
|반복 | * 연산자를 이용하여 시퀸스 자료를 여러번 반복하여 새로운 시퀸스 자료를 생성|
|멤버 체크| in 키워드를 사용하여 특정값이 시퀸스 자료의 요소로 속해 있는지 확인 가능함 |
|크기 정보| len()을 사용하여 시퀸스 자료의 크기를 알 수 있음. 문자열은 문자의 수, 리스트와 튜플은 맴버의 수가 됨|

````python
strdata = 'abcde'
listdata = [1,[2,3],"안녕"]
tupledata = (100,200,300)

for char in strdata:
    print(char, end=" ")
print("\nstrdata 의 순서 출력입니다.")
print()

for i in range(len(listdata)):
    print(listdata[i], end=" ")
print("\nlistdata의 순서 출력입니다")

print()
print(tupledata[1])
try :     
    tupledata[1] = 0
except:
    print("tuple는 수정 불가능합니다.")
````

## 28.시퀸스 자료 슬라이싱 이해

시퀸스 자료에서 일정 범위에 해당하는 부분을 취하는 방법  
구간은 [start : end : step]로 정의됨.   

start : 슬라이싱 범위의 시작, 슬라이싱에 포함함. 생략할 시 0부터 시작한다는 것을 의미함. 
end   : 슬라이싱 범위의 끝, 슬라이싱에서 포함하지 않음. 생략할 시 시퀸스 자료의 마지막까지 포함하는걸 의미함
step  : 자료를 취하는 간격 인자를 생략할 시 기본값으로 1을 적용함. 음수를 사용할 시 역순으로 간격을 취하는 것을 의미함. 

````python
strdata = "Time is my money"

print(strdata[1:5]) # 내부의 특정 범위
print(strdata[:7]) # 처음부터 특정 위치까지
print(strdata[1:]) # 특정 위치부터 마지막까지
print(strdata[-5:]) # 역순으로 특정 위치부터 마지막까지
print(strdata[:-5]) # 처음부터 역순으로 특정 위치까지
print(strdata[::2]) # 2칸 단위로 슬라이싱
print(strdata[::-1]) # 역순 출력

````

## 34. 문자열 포메팅 이해
문자열 포메팅은 변하는 값을 포함하는 문자열을 포함하기 위해 하나의 양식으로 문자열을 만드는 것.  
포멧 문자열의 경우 크게 3가지 방식으로 구현됨. 
1. C언어 타입. %s, %c 등을 통해 원하는 타입을 지정하여 포메팅을 수행
2. "{}과 {}를 원함".format(a,b) 순서대로 변수를 
3. f"{a}와 {b}를 원함" 문자열에 변수를 바로 전달함.  -> 현재 표준

포메팅의 경우 다음과 같은 기능을 포함하여 다양한 기능이 있음.  
1. 변수명 뒤에 = 를 추가하여 변수명과 변수 값을 동시에 출력  
    var = 10  
    print(f"{var=}") # 출력 var=10

2. 소수점 자리수 지정 :.nf
    print(f"{pi:.2f}")

3. 천단위 콤마 :,  
    var3 = 1000000  
    print(f"{var1:,}")  

4. 정렬  :<n,:>n,:^n 각각 n칸에 대해 좌 우 중앙  
    var4 = 10  
    print(f"{var4:<10}")  
    print(f"{var4:>10}")  
    print(f"{var4:^10}")  


````python
import math
var = 10  
print(f"{var=}") # 출력 var=10

pi =math.pi
print(f"{pi:.4f}") 

var1 = 1000000
print(f"{var1:,}")

var4 = 10
print(f"{var4:<10}")
print(f"{var4:>10}")
print(f"{var4:^10}")

````

## 36. 리스트 이해
리스트는 []로 표시하며 []안의 요소를 콤마로 구분하여 순서있게 나열함.  
리스트의 요소는 임의의 객체가 가능하며 임의의 객체는 숫자, 문제 문자열, 리스트, 튜플, 사전 등 파이썬 자료형과 함수, 클래스 등을 의미함.
리스트는 인덱싱을 활용하여 해당 인덱스의 값을 변경 가능함. 

````python
list1 = [1,2,3,4,5]
list2 = ["a", "b","c"]
list3 = [1,"a","abc",list1, list2]

print(list1)
list1[0] = 6 # 0번 인덱스의 값을 1에서 6으로 변경
print(list1)
print(list2)
print(list3)

def myfunc():
    print("안녕하세요")

list4 = [1,2, myfunc] # 함수 myfunc를 저장하여 원할 때 실행 가능함
list4[2]()  # 인자를 넣지 않고 list4의 2번 인덱스의 함수를 실행

list5 = [1,2,myfunc()] # 함수 myfunc의 반환값을 리스트에 저장. 이 과정에서 함수 myfunc가 실행됨. 
print(list5)
````
