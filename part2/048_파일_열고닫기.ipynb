## 48. 파일 열고 닫기
파일은 텍스트 파일과 바이너리 파일 두가지 종류가 있음.   
텍스트 파일은 사람이 읽을 수 있는 글자로 저장된 파일이고,    
바이너리 파일은 커퓨터가 읽고 이해할 수 있는 이진 데이터를 기록한 파일로 이미지 파일, exe 파일 등이 있음.    
파일을 사용하기 위해서는 open(파일이름,모드)를 사용하여 열 수 있으며 성공적으로 오픈하면 파일 객체를 반환함.  
파일이름은 파일이 존재하는 상대경로 혹은 절대 경로를 의미하며 모드는 다음 표와 같음   
|모드 | 설명|
|-----|-----|
|r or rt| 텍스트 모드로 읽기|
|w or wt| 텍스트 모드로 쓰기|
|a or at| 텍스트 모드로 파일 마지막에 추가하기|
|rb | 바이너리 모드로 읽기|
|wb| 바이너리 모드로 쓰기|
|ab | 바이너리 모드로 마지막에 추가하기|

파일을 다루는 것을 마치면 자원 관리를 위해 파일 객체의 close()를 이용하여 오픈한 파일을 닫아야 함. 

```python
## 이 코드에서 사용한 파일은 파일명만 맞춰 폴더에 넣는걸 권장함. 
# 텍스트 파일을 확인하는 방법
try :
    f1 = open("part2/text.txt", 'r', encoding='utf-8')
    data = f1.read()
    print(data)
    f1.close()
except : 
    print("파일 호출을 실패했습니다")

#바이너리 파일 확인 
try:
    f2 = open("part2/image.jpg", "rb")
    content = f2.read(20) # 앞의 20바이트만 읽기
    print(f"파일 객체 정보: {f2}")
    print(f"실제 바이너리 데이터(일부): {content}")
    f2.close()
except FileNotFoundError:
    print("파일 경로를 다시 확인해주세요!")
```
