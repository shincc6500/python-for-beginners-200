## 16. 정수형 자료 이해하기

파이썬에서 정수형 상수의 범위는 메모리가 허용하는 범위에서 지원 범위가 정의됨. 

```python
int_data =10
bin_data = 0b10
oct_data =0o10
hex_data= 0x10
long_data = 1234567890123456789

print(int_data)
print(bin_data)
print(oct_data)
print(hex_data)
print(long_data)
```

# 17. 실수형 자료 이해하기

실수는 소수로 나타낼 수 있는 유리수와 원주율과 같은 무리수로 구성된 수 집합
실수형 자료형은 정수라도 실수형으로 취급됨. 

실수형 상수를 표현할 때 10의 거듭제곱을 e로 표현할 수 있음

```python
f1 = 1.0
f2 = 3.14
f3 = 1.56e3
f4 = -5e-2

print(f1)
print(f2)
print(f3)
print(f4)
```

## 18. 복소수형 자료 이해하기
복소수형 사수는 실수부 + 허수부로 이루어진 수로 일반적인 수학에서는 허수부를 i로 사용하지만 파이썬에서는 j를 사용하여 표현함.   
복소수에서 실수부만 출력하려면 변수 뒤에 .real을, 복소수만 출력하려면 .imag를 붙이면 됨  
또한 complex()를 이용하여 복소수형 상수를 구현 가능함.   

```python
c1 = 1 +7j
print(c1)
print(c1.real)
print(c1.imag)

c2 = complex(2,3)
print(c2)
```
