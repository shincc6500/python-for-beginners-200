## 55. 예외처리
프로그램 작성시 예상하지 못한 오류 발생하는 코드가 있을 수 있음.   
이러한 오류가 발생시 더이상 진행 될 수 없는 상태가 됨.   
이러한 오류가 발생하더라도 프로그램을 중단시키지 않고  
예외에 대한 적절한 처리를 하여 프로그램을 계속 진행시킬 수 있는 구문이 try except 구문임  
try 영역에 대해서 코드를 실행하다 오류가 발생하면 except 영역의 코드를 수행하고 코드를 계속 실행하기 위한 구문임. 

```python
try : 
    print("안녕하세요")
    print(param)
except:
    print("예외가 발생했습니다. ")
```

## 56. 예외처리 2
로직을 수행할 때 오류 상황이 아닐 경우에만 특정 코드를 수행하려면 try except 뒤에 else를 붙여 사용할 수 있음. 

```python
try : 
    print("안녕하세요")
    print(param)
except:
    print("예외가 발생했습니다. ")
else : 
    print("예외가 발생하지 않았습니다. ")

param = "예외 없음"
try : 
    print("안녕하세요")
    print(param)
except:
    print("예외가 발생했습니다. ")
else : 
    print("예외가 발생하지 않았습니다. ")
```

## 57. 예외처리 3   
오류 발생 유무와 상관없이 특정 코드를 무조건 실행시키려면 finally를 사용하면 됨.    
파일 열고 닫기 등의 메모리 점유율이 높은 기능을 사용할 경우 오류가 발생하여 종료 코드가 실행되지 않는 경우 문제가 될 수 있는 경우 주로 사용함.   

```python
try : 
    print("안녕하세요")
    print(param)
except:
    print("예외가 발생했습니다. ")
else : 
    print("예외가 발생하지 않았습니다. ")
finally:
    print("생략하면 안되는 코드")

param = "예외 없음"
try : 
    print("안녕하세요")
    print(param)
except:
    print("예외가 발생했습니다. ")
else : 
    print("예외가 발생하지 않았습니다. ")
finally:
    print("생략하면 안되는 코드")
```

```python
try:
    print(param3)
except Exception as e:
    print(e)
```

## 59. 예외 처리 5
코드에서 특정 예외 상황이 발생한것만 특정 코드를 처리하기 위해서는 except 뒤에 특정 예외를 명시해주면 됨. 


```python
import time
count =1
try :
    while True:
        print(count)
        count +=1
        time.sleep(0.5)
except KeyboardInterrupt: # Crtl+C 가 입력되면 발생하는 오류
    print("사용자에 의해 프로그램이 중단되었습니다.")
```
