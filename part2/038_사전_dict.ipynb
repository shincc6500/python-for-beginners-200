## 사전(dict) 이해하기
dict는 키(key)와 값(value) 쌍을 하나의 요소로 하는 순서가 없는 집합  
이로 인해 dict는 시퀸스 자료형이 아니라서 인덱싱으로 접근이 불가능함.   
dict의 키와 값은 각각 임의의 객체가 될 수 있음  

```python
dict1 = {'a':1,'b':2,'c':3}
print(dict1['a']) #특정 키값의 value 출력
dict1['d'] = 4 #dict key:value 추가
print(dict1) 
dict1['b']=7 #dict value 수정
print(dict1) 
print(len(dict1)) # dict의 크기 
```
