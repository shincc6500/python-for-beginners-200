## 49. 클래스 이해
클래스는 프로그래머가 지정한 이름으로 만든 하나의 독립된 공간이며, 이름 공간이라 부름. 
클래스를 구성하는 주요 요소는 변수 역할을 하는 클래스 맴버와, 함수와 동일한 역할을 하는 클래스 메소드가 있음. 

클래스는 다음과 같이 정의함. 
```python 
class 클래스 명:
    클래스 맴버 정의
    클래스 메소드 정의 

클래스 멤버와 클래스 메소드는 클래스명.멤버명 or 클래스명.메소드명 의 형식으로 호출함. 

````python
class MyClass:
    var = "안녕하세요"
    def sayHello(self):
        print(self.var)

obj = MyClass()
print(obj.var)
obj.sayHello()
````

## 50. 클래스 멤버와 인스턴스 멤버 이해
클래스에서 선언되는 변수는 클래스 멤버와 인스턴스 멤버로 구분할 수 있음. 

클래스 멤버는 클래스 메소드 바깥에서 선언되고, 모든 인스턴스가 '공유'하는 변수임. 클래스가 메모리에 로드 시 생성 - 싱글톤 패턴과 유사  
인스턴스 맴버는 클래스 내에서 self와 같이 선언되는 변수로 각 객체가 다른 값을 가지게 됨. 

````python
class MyClass:
    var = "안녕하세요!!"
    def sayHello(self):
        param1="안녕"
        self.param2="하이"
        print(param1)
        print(self.var)

obj = MyClass()
print(obj.var)
obj.sayHello()
try:
    print(obj.param1) #param1의 경우 인스턴스 변수, sayHello() 메서드 동작과 함께 생성함. 
except:
    print("오류 발생")

obj2 = MyClass()
print(obj2.var)
````

## 51. 클래스 메소드 이해
메소드는 클래스 내에서 정의되는 함수임. 
메소드는 첫번째 인자가 반드시 클래스의 인스턴스 객체를 지시하는 self여야 함. 
별도의 인자가 필요하다면 self 뒤에 추가하면 됨. 

```python
class 클래스명:
    def 메소드명(self):
        동작 코드
```

````python
class MyClass:
    def sayHello(self):
        print("안녕하세요")
    
    def sayBye(self, name):
        print(f"{name} 다음에 보자")

obj =MyClass()
obj.sayHello()
obj.sayBye("철수")
````

## 52. 클래스 생성자 이해
클래스 생성자는 인스턴스 객체가 생성될 때 자동적으로 호출되는 메소드.  
클래스 생성자는 다음과 같은 이름을 가짐 
```python 
def __init__(self, *args):
```
이 생성자는 사실상 인스턴스 생성과 동일한 기능이므로, 인스턴스 객체 생성할 때 반드시 생성자의 인자에 값을 전달해 줘야 함. 


````python
class MyClass:
    def __init__(self, var):
        self.var = var
        print("Myclass 객체가 생성되었습니다")

obj= MyClass("안녕")
print(obj.var)
obj2 = MyClass("안녕하세요")
print(obj2.var)
````

## 53. 클래스 소멸자 이해
클래스 소멸자는 클래스 인스턴스 객체가 메모리에서 제거될때 자동적으로 호출되는 클래스 메소드
클래스 소멸자는 다음과 같은 이름을 가짐
```python 
__del__(self):
```
이를 통해 "del <인스턴스 객체> " 코드를 통해 인스턴스 객체를 메모리에서 제거 가능함. 

````python
class MyClass:
    def __del__(self):
        print('MyClass 인스턴스 객체가 메모리에서 제거됩니다.')

obj = MyClass()
del obj

try :
    del obj
except:
    print("이미 삭제됬습니다.")
````

## 54. 클래스 상속 이해하기. 
클래스는 어떤 클래스가 가지고 있는 모든 멤버나 메소드를 다른 클래스가 모두 사용할 수 있도록 하는 상속이라는 기능이 있음.  
상속을 해 주는 클래스를 부모클래스 or 슈퍼클래스라고 부르고,  
상속을 받는 클래스를 자식클래스 or 서브클래스 라고 부름.  

상속은 다음과 같은 형식으로 정의하며, 부모클래스를 여럿 사용할 경우 다중 상속이라고 함.   
단 다중상속의 경우 사용시 코드 관리가 어려울 수 있어서 권장하지 않는 방식임.   
```python 
class 자식클래스(부모클래스)
```

````python
class Add:
    def add(self,n1,n2):
        return n1+n2
class Calculator(Add):
    def sub(self, n1, n2):
        return n1-n2
    
obj = Calculator()
print(obj.add(1,2))
print(obj.sub(1,2))
````
