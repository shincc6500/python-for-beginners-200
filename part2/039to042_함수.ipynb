## 39. 함수 이해하기
함수란 특정 목적을 가진 코드의 집합으로, 독립적으로 호출 될 수 있음.

함수는 기본적으로 다음과 같은 형식으로 선언함. 조건에 따라 인자나 결과값 혹은 둘 모두를 생략할 수 있음.   
이때 매개변수에 등호를 사용하여 기본값을 정의할 수 있으나, 기본값이 정해진 매개변수가 기본값이 없는 매개변수 앞에 올 경우 오류가 발생함.

```python
def 함수이름(매개변수1, 매개변수2):
    실행할 코드
    return 결과값  # 결과가 없다면 return 생략 가능
```

함소를 호출 할 때는 다음과 같은 방식을 사용함. 인자값의 경우 함수 선언시 조건에 의존함. 
```python
# 반환값을 변수에 할당하기 위한 경우
var = function(인자값1, 인자값2, ......)

# 반환값이 없거나 사용하지 않는 경우 
function(인자값1, 인자값2,....)
```


````python
# 함수 예제
def add_number(n1,n2):
    ret = n1+n2
    print(ret)
    return ret

def add_txt(t1, t2):
    print(t1+t2)

ans = add_number(10,15)
print("함수 실행 과정에서 출력된 값")
print(ans, "\n함수로 할당된 변수를 출력")


text1 = "대한민국 "
text2 = "만세!"
add_txt(text1,text2)
````

## 40. 함수 인자 이해. 
함수가 요구하는 값을 인자라고 한다. 
이러한 인자를 입력하기 위해서는 다음과 같은 방법이 있다. 
1. 실제값을 대입하여, 인자 순서에 대응하여 값을 입력
2. 기본 인자를 이용하여 값을 전달. - 디폴트값을 지정하여 특정 값을 입력
3. 인자 이름을 이용하여 값을 전달. - 함수 호출할 때 매개변수 명에 인자값을 할당하여 원하는 변수에 원하는 값을 입력. 
4. 가변 인자 튜플을 이용하여 값을 전달. - 정해지지 않은 수의 인자 목록을 받을때 사용
5. 미정 키워드 인자 사전을 이용하여 값을 전달. - 주로 정해지지 않은 수의 속성/설정을 처리할때 주로 사용. 

가변인자란?
"*" 혹은 "**"를 변수 앞에 붙여 지정되지 않은 수의 인자를 튜플 혹은 dict로 묶어 처리하는 방식  
*args : 지정되지 않은 수의 인자를 튜플로 묶어서 처리함. 이때 튜플은 시퀸스 자료형이므로 순서가 중요함.   
**kwargs : 지정되지 않은 수의 인자의 이름:값 쌍을 dict로 묶어서 처리함.  


````python
def add_txt(t1, 
            t2="파이썬"):
    print(t1+":"+t2)

add_txt("베스트") # 기본 인자를 이용하여 값을 전달
add_txt(t2="대한민국", t1="1등") # 인자 이름을 이용하여 값을 전달

def func1(*args): # 가변 인자 튜플을 이용한 값 전달
    print(args)

def func2(**kargs): # 미정 키워드 인자 사전을 이용한 값 전달    
    for key, value in kargs.items():
        print(f"{key}: {value}")

func1() # 빈 튜플이 출력
func1(3,5,1,5) # (3,5,1,5)가 출력

func2(name="Alice", age=25, city="Seoul")
````

## 41. 지역변수와 전역변수의 이해
변수의 유효한 범위를 기준으로 지역변수와 전역변수를 구분함.   
- 지역변수 : 함수 내부에서만 유효하며, 함수가 실행될 때 생성되고 종료되면 같이 소멸함. 
- 전역변수 : 함수 밖에서 선언되는 함수로 프로그램 전반에서 유효함. 

지역변수와 전역변수가 동일한 이름을 가지더라도 이 둘은 독립된 값임으로 함수의 배개변수 사용시 이 특성이 가장 크게 나타남.  
함수 내부에서 전역 변수에 접근해서 수정하기 위해서는 global을 사용해야 함. 

````python
intvar = 10
strvar = "전역변수"

def func1():
    strvar = "지역변수"
    print(strvar)

def func2(intvar):
    print(intvar)

def func3():
    global intvar
    intvar +=10
    print(intvar)

print(strvar) # 전역변수 출력
func1() # 지역변수 출력

print(intvar) # 전역변수 출력
func2(5) # 인자 = 지역변수/ 전역변수와 별개
func3() # 함수 내에서 전역변수 수정
print(intvar) # 전역변수 출력
````

## 42. 함수 리턴값 이해
함수는 return 키워드를 이용하여 값을 반환하고 함수를 종료할 수 있음.   
리턴값이 다수일 경우 튜플로 리턴값을 만들어 리턴하게 됨.   
이때 리턴값을 하나씩 변수에 할당 할 경우 기본적으로 변수의 수와 리턴값의 수가 같아야 함.   
변수를 하나하나 할당 하지 않으려면 다음과 같은 조치가 필요함.   
1. 필요 없는 값은 _(언더바)로 무시
2. 가변 인자 "*"를 이용하여 리스트로 처리(가변인자는 위치에 따라 튜플과 리스트가 가변적임. )

````python
def reverse(x,y,z):
    return z,y,x
ret = reverse(1,2,3)
print(ret)

try:
    r1,r2 = reverse(4,5,6)
    print(r1,r2)
except:
    print("할당 변수와 리턴값의 수를 맞춰 주세요")

r1,r2,_ =reverse(4,5,6)
print(r1,r2)

x1, *args = reverse(7,8,9)
print(x1)
print(args)
````
