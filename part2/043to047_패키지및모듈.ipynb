## 44. 파이썬 패키지 이해
파이썬 패키지는 파이썬 모듈을 계층적인 디렉터리 형태로 구성한 것임.  
디렉터리를 패키지로 인식하기 위해서는 각 디렉터리마다 "__init__.py"파일이 있어야 합니다. 이때 해당 파일은 비어있거나 버전 정보 정도만 있어도 충분함.  

## 45. 파이썬 모듈 임포트 이해
모듈 임포트는 다음과 같은 구성을 가짐. 

```python 
import 모듈명
import 패키지명.모듈명 # 사용할때 패키지명.모듈명 구조로 호출해야 함. 
from 패키지명 import 모듈명 # 사용할 때 모듈명만 호출해도 됨
import 패키지명.모듈명 as 별칭 # 모듈명이 길어질 경우 간단하게 작성하기 위해서 별칭을 사용하여 호출하도록 정의할 수 있음.
```

패키지 내부의 특정 모듈을 호출하기 위해서는 두번째 혹은 3번째 방식을 사용해야 함.  
이때 패키지명.모듈명은 주로 서로 다른 패키지 내부에 있는 동일한 모듈명으로 인한 오류를 방지하기 위해 사용하는 방식임. 
이 경우 호출명이 길어지기 때문에 as를 사용하여 모듈을 별칭으로 호출하는 경우가 많음. 

또한 존재하지 않는 모듈을 임포트하면 ImportError가 발생함.
