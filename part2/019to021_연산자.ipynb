## 20. 사칙 연산자 이해
사칙연산은 일반적인 사칙 연산 순서 기준으로 적용
또한 거듭제곱은 사칙연산보다 우선순위로 처리 됨. 
**은 거듭제곱을 
//은 나머지를 출력으로 내놓음

```python
a=2
b=4
ret1 = a+b
ret2 = a-b
ret3 = a*b
ret4 = a/b
ret5 = a**b
ret6 = a//b
ret7 = a+a*b/a
ret8 = (a+b)*(a-b)
ret9 = a*b**a


for i in range(1, 10):  
    var_name = f"ret{i}"      # "ret1", "ret2"... 문자열 생성
    print(var_name, "=",locals()[var_name]) # 변수 명부(locals)에서 해당 이름의 값을 꺼냄 이 방식은 기초적인 테스트 용도로만 사용할 것!!!!
```

## 연산자 축약 이해하기
사칙 연산을 수행하고 그 결과를 다시 변수 a에 대입할 때 사용하는 방식   
연산자와 등호의 조합으로 구현됨. 

```python
a = 0
a +=1
print(a)
a-=5
print(a)
a *=2
print(a)
a/=4
print(a)
```
