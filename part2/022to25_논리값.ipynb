## 22. True와 False

True == 1

## 23. 관계 연산자의 이해
관계 연산자는 두개의 값을 서로 비교하여 참인지 거짓인지 판단할 때 사용되는 연산자. 

|관계연산자 | 의미|
|----------|--------|
|A==B | A와 B가 같으면 참|
|A!=B | A와 B가 다르면 참|
|A<B | A가 B보다 작으면 참|
|A<=B | A가 B보다 작거나 같으면 참|
|A>B | A가 B보다 크면 참|
|A>=B | A가 B보다 크거나 같으면 참|


```python
x=1
y=2
str1 = "abc"
str2 = "python"

print(x==y) # 둘이 다르므로 false
print(x!=y) # 둘이 다르므로 True
print(str1==str2) # 둘이 다르므로 false
print(str2 =="python") #str2에 "python"이 할당되어 있으므로 참
print(str1<str2)  #문자열의 크기 비교는 문자열의 사전 순서로 비교하여 참
```

## 24. 논리 연산자의 이해  
논리 연산자는 참, 거짓으로 되어 있는 두개의 값을 비교하여 참 또는 거짓으로 결과를 내놓는 연산자

|논리 연산자| 의미 |
|---------|-------|
|A and B| A와 B모두가 참이면 참|
|A or B| A,B 둘중 하나 이상이 참이면 참|
|notA | A의 논리값의 반대|

```python
bool1 = True
bool2 = False
bool3 = True
bool4 = False

print(bool1 and bool2)
print(bool1 and bool3)
print(bool2 or bool3)
print(bool2 or bool4)
print( not bool1)
print( not bool2)
```

## 25. 비트 연산자 이해 

1비트는 0또는 1로만 표현될 수 있는 데이터 단위.
컴퓨터의 1byte는 8bit => 2^8=256개의 표현이 가능한 값  
컴퓨터는 이진연산이 기본
파이썬의 사용 가능한 비트 연산자는 다음과 같음. 

|비트 연산자 |의미|
|-------|------|
| A&B | A와 B사이의 and 연산 수행|
| A|B | A와 B사이의 or 연산 수행|
| A^B | A와 B사이의 xor 연산 수행|
| ~A | A의 보수를 계산. 비트 반전|
| A>>n | A의 모든 비트를 n만큼 오른쪽으로 시프트|
| A<<n | A의 모든 비트를 n만큼 왼쪽으로 시프트|

```python
bit1 = 0x61
bit2 = 0x62

print(hex(bit1&bit2))
print(hex(bit1|bit2))
print(hex(bit1^bit2))
print(hex(bit1>>1))
print(hex(bit1<<2))
print(hex(~bit1))
```
