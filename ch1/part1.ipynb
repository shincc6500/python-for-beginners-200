자료형 및 자료형 출력  
파이썬에는 5가지 기본 자료형이 있으며 자료형을 통해 여러 종류의 데이터를 구분하기 위한 분류임.   
1. 수치형 자료  
    1) 정수형 상수 - 정수  
    2) 실수형 상수 - 유리수와 무리수를 포함한 실수   
    3) 복소수형 상수 - 실수부 + 허수부   
2. 문자열 자료    
    - 한글자 이상의 문자나 숫자, 기호로 구성된 자료   
    이때 문자열 자료"1"과 수치형 자료 1은 다른 값임  
3. 리스트 자료 (list)  
    - [] 안에 임의의 객체를 순서 있게 나열한 자료형  
4. 튜플 자료 (tuple)  
    - ()를 사용하여 리스트와 유사하게 임의의 객체를 순서있게 나열한 자료형  
    리스트와는 다르게 요소값 변경이 안됨.   
5. 사전 자료(dict)  
    -{}를 사용하여 생성하는 키 : 값 쌍이 소요로 구성된 순서가 없는 자료형임.  
    순서가 없어 인덱스 대신 키 값을 사용하여 접근   

````python
int_data =1
float_data = 3.14
comlplex_data = 1+5j

str_data1 = 'I love Python' # ''를 사용하여 문자열 자료형 생성
str_data2 = "안녕하세요 " # ""를 사용하여 문자열 자료형 생성

list_data = [1,2,3]
tuple_data = (1,2,3)
dict_data = {0:'False', 
             1:"True"} # key는 0,1 value는 'False', "True"

````

## 출력 명령어(print)

### print는 인자를 다음과 같이 받음   

- print(value, sep, end, file, flush)  

    value : 출력할 값들, 콤마를 사용하며 무제한으로 나열 가능  

    sep : 구분자, 여러개의 인자를 입력했을 때 각 인자를 구분하기 위한 값 기본값 " "(한칸 띄어쓰기)  

    end : 마지막에 입력 처리할 내용 기본값 "\n"(줄바꿈)  

    file : 어디에 출력할 것인가를 결정. 기본값은 sys.stdout (터미널)이나 원하는 경우 파일에 직접 출력 가능함. 대표적으로 log 파일이 있음.   
    
    flush : 버퍼에서 출력 마무리 되고 나서 출력할것인지 여부를 결정. 기본값 False. True를 사용할 경우 타이핑하는 것과 같이 출력 되는 효과가 나옴.   

````python
print(int_data) 
print(str_data1)

print(list_data) #리스트 전부를 출력
print(list_data[1]) #리스트 중 1번 위치의 값을 출력/ 리스트는 0번 부터 시작

print(dict_data) # dict data 전부를 출력
print(dict_data[1]) #dict_data에서 key 값이 1인 value를 출력

````

````python
list_data = ["a", "b", "c"]
if "a" in list_data:
    print("a가 list_data에 있습니다")
    print(list_data)
else:
    print("a가 list_data에 없습니다")
````

if 문은 조건에 따라 수행해야 하는 코드를 다르게 하기 위한 코드임. 
단일 조건만 비교할 경우 if-else 구조, 다중 조건을 비교할 경우 if-elif-else 구조를 사용함. 
만약 조건이 참인 경우만 특정 코드를 실행하는 코드의 경우 else를 생략 가능함. 

if문의 기본 구조는 다음과 같음. 
```python 
if 조건:
    실행코드 
elif 조건:
    실행코드
else:
    실행코드
```

````python
x=1
y=2

if x>y:
    print("x가 y 보다 큽니다")
elif x<y:
    print("x가 y보다 작습니다")
else:
    print("x는 y와 같습니다")
````

반목문(for)  
for문은 특정 범위의 자료나 객체에 대해 처음부터 끝까지 하나씩 추출하여 특정 코드를 반복적으로 수행하는 코드     
for 문의 기본 구조는 다음과 같으며 범위로 올수 있는 값은 문자열, 리스트, 튜플, dict, range(), 그 외 반복 가능한 객체 가 있음. 

```python 
for 변수 in 범위:
    반복으로 실행할 코드
```
 
 이러한 반복문을 추가적으로 제어하기 위해 다음과 같은 명령어가 있음. 
 - continue : 이번 실행 코드를 종료하고 다음 반복을 실행
 - break : 반복문에서 탈출
 - else : 반복문이 끝까지 수행 됬을때 실행하는 코드 




````python
#기본적인 반복문 
scope = [1,2,3,4,5]
for x in scope:
    print(x)

# continue와 break를 사용
for x in scope:
    print(x)
    if x<3:
        continue
    else :
        break
    print("continue로 인해 생략됩니다")
````
